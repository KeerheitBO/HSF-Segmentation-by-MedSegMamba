```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import nibabel as nib
import pandas as pd
from medpy.metric.binary import dc, hd95, assd
from skimage import measure
from tqdm import tqdm
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from scipy.ndimage import binary_erosion, binary_dilation

from models.hilbert import decode, encode
from models.MedSegMamba import MedSegMamba
from sklearn.model_selection import train_test_split
from torch.utils.data import Dataset, DataLoader
import torch.nn.functional as F

image_folder = '/content/drive/MyDrive/4460_Project/hsf_otherpaireddata_patch_pt'
label_folder = '/content/drive/MyDrive/4460_Project/hsf_otherpaireddata_patch_label_pt'

# Extrat all the file names
image_files = sorted([f for f in os.listdir(image_folder) if f.endswith('.pt')])
label_files = sorted([f for f in os.listdir(label_folder) if f.endswith('.pt')])

# Match the names
def get_base_name(name):
    return name.split('_hippo')[0]  # Use

# Build label_dict for pairing
label_dict = {get_base_name(f): os.path.join(label_folder, f) for f in label_files}

image_paths = []
label_paths = []

for img_file in image_files:
    base = get_base_name(img_file)
    if base in label_dict:
        image_paths.append(os.path.join(image_folder, img_file))
        label_paths.append(label_dict[base])
    else:
        print(f'⚠️ Label not found for: {img_file}')


assert len(image_paths) == len(label_paths), "Mismatch between images and labels!"

# Split
train_imgs, temp_imgs, train_labels, temp_labels = train_test_split(
    image_paths, label_paths, test_size=0.3, random_state=42
)

val_imgs, test_imgs, val_labels, test_labels = train_test_split(
    temp_imgs, temp_labels, test_size=0.5, random_state=42
)

print(f'Train: {len(train_imgs)}, Val: {len(val_imgs)}, Test: {len(test_imgs)}')

# Create custom Datasets
class MRISegmentationDataset(Dataset):
    def __init__(self, image_paths, label_paths):
        self.image_paths = image_paths
        self.label_paths = label_paths

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image = torch.load(self.image_paths[idx])
        label = torch.load(self.label_paths[idx])

        # Add channel dimension to image
        if image.ndim == 3:
            image = image.unsqueeze(0)  # image -> [1, 96, 96, 96]

        # Label stays 3D, but must be long dtype for loss
        if label.ndim == 3:
            label = label.unsqueeze(0)  # label shape -> [1, 96, 96, 96]

        return image, label

# Datasets
train_dataset = MRISegmentationDataset(train_imgs, train_labels)
val_dataset = MRISegmentationDataset(val_imgs, val_labels)
test_dataset = MRISegmentationDataset(test_imgs, test_labels)

# DataLoaders
train_loader = DataLoader(train_dataset, batch_size=1, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=1, shuffle=False, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False, num_workers=2)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts
import time
import numpy as np
from tqdm import tqdm
import torch.nn.functional as F

class DiceLoss(nn.Module):
    def __init__(self, smooth=1.0, ignore_background=True):
        super().__init__()
        self.smooth = smooth
        self.ignore_background = ignore_background

    def forward(self, inputs, targets):

        if targets.dim() == 5 and targets.shape[1] == 1:
            targets = targets[:, 0]

        num_classes = inputs.shape[1]
        targets_one_hot = F.one_hot(targets.long(), num_classes).permute(0, 4, 1, 2, 3).float()

        dims = (0, 2, 3, 4)
        intersection = torch.sum(inputs * targets_one_hot, dims)
        cardinality = torch.sum(inputs + targets_one_hot, dims)
        dice_score = (2. * intersection + self.smooth) / (cardinality + self.smooth)

        if self.ignore_background:
            dice_score = dice_score[1:]

        return 1 - torch.mean(dice_score)

class CombinedDiceWCE(nn.Module):
    def __init__(self, class_weights, dice_weight=0.4, ce_weight=0.6):
        super().__init__()
        self.dice = DiceLoss(ignore_background=True)
        self.ce = nn.NLLLoss(weight=class_weights)
        self.dice_weight = dice_weight
        self.ce_weight = ce_weight

    def forward(self, inputs, targets):
        log_inputs = torch.log(inputs + 1e-8)
        targets = targets.squeeze(1) if targets.dim() == 5 else targets
        loss_dice = self.dice(inputs, targets)
        loss_ce = self.ce(log_inputs, targets)
        return self.dice_weight * loss_dice + self.ce_weight * loss_ce

# Evaluation metrics
def calculate_metrics(outputs, targets, num_classes):
    preds = torch.argmax(outputs, dim=1)

    # Calculate Dice scores for each class
    dice_scores = []
    vs_scores = []  # Volume Similarity scores

    for cls in range(1, num_classes):  # Skip background class 0
        pred_cls = (preds == cls).float()
        target_cls = (targets == cls).float()

        # Skip if class not present in ground truth
        if target_cls.sum() == 0:
            continue

        # Dice calculation
        intersection = (pred_cls * target_cls).sum().item()
        dice_score = (2. * intersection) / (pred_cls.sum().item() + target_cls.sum().item() + 1e-8)
        dice_scores.append(dice_score)

        # Volume Similarity calculation
        vs_score = 1 - abs(pred_cls.sum().item() - target_cls.sum().item()) / (pred_cls.sum().item() + target_cls.sum().item() + 1e-8)
        vs_scores.append(vs_score)

    avg_dice = np.mean(dice_scores) if dice_scores else 0
    avg_vs = np.mean(vs_scores) if vs_scores else 0

    return {
        'dice': avg_dice,
        'volume_similarity': avg_vs
    }

# Training function
def train_epoch(model, train_loader, optimizer, criterion, device, epoch):
    model.train()
    epoch_loss = 0
    total_batches = len(train_loader)
    patches_per_scan = 27  # As mentioned in the paper: 3 steps per dimension

    optimizer.zero_grad()  # Zero gradients at start
    accumulated_loss = 0

    # Use tqdm for progress tracking
    progress_bar = tqdm(enumerate(train_loader), total=total_batches,
                        desc=f'Training Epoch {epoch+1}')

    for batch_idx, (data, target) in progress_bar:
        data, target = data.to(device), target.to(device)

        # Forward pass
        output = model(data)
        loss = criterion(output, target)

        # Backpropagation
        loss.backward()
        accumulated_loss += loss.item()

        # Update weights after accumulating gradients for all patches of a scan
        if (batch_idx + 1) % patches_per_scan == 0 or (batch_idx + 1) == total_batches:
            optimizer.step()
            optimizer.zero_grad()

            # Update progress bar
            progress_bar.set_postfix({'loss': accumulated_loss / min(patches_per_scan, batch_idx % patches_per_scan + 1)})
            epoch_loss += accumulated_loss
            accumulated_loss = 0

    return epoch_loss / total_batches

# Validation function
def validate(model, val_loader, criterion, device, epoch, num_classes):
    model.eval()
    val_loss = 0
    metrics = {'dice': 0, 'volume_similarity': 0}
    num_batches = len(val_loader)

    with torch.no_grad():
        for data, target in tqdm(val_loader, desc='Validation'):
            data, target = data.to(device), target.to(device)

            output = model(data)
            loss = criterion(output, target)
            val_loss += loss.item()

            # Calculate metrics
            batch_metrics = calculate_metrics(output, target, num_classes)
            metrics['dice'] += batch_metrics['dice']
            metrics['volume_similarity'] += batch_metrics['volume_similarity']

    # Average metrics
    val_loss /= num_batches
    metrics['dice'] /= num_batches
    metrics['volume_similarity'] /= num_batches

    return val_loss, metrics

# Testing function
def test(model, test_loader, device, num_classes):
    model.eval()
    metrics = {'dice': 0, 'volume_similarity': 0}
    num_batches = len(test_loader)

    with torch.no_grad():
        for data, target in tqdm(test_loader, desc='Testing'):
            data, target = data.to(device), target.to(device)

            output = model(data)

            # Calculate metrics
            batch_metrics = calculate_metrics(output, target, num_classes)
            metrics['dice'] += batch_metrics['dice']
            metrics['volume_similarity'] += batch_metrics['volume_similarity']

    # Average metrics
    metrics['dice'] /= num_batches
    metrics['volume_similarity'] /= num_batches

    return metrics

# Main training loop
def train_and_evaluate(model, train_loader, val_loader, test_loader, num_classes,
                       epochs=15, learning_rate=1e-4, device='cuda', save_path='best_model.pth'):

    optimizer = optim.AdamW(model.parameters(), lr=learning_rate)
    metrics_log = {
    "epoch": [],
    "train_loss": [],
    "val_loss": [],
    "val_dice": [],
    "val_vs": []
}

    best_val_dice = 0
    class_weights = torch.ones(num_classes, dtype=torch.float)
    class_weights[0] = 0.001  # lower weight for background
    class_weights = class_weights.to(device)
    criterion = CombinedDiceWCE(class_weights)


    for epoch in range(epochs):

        start_time = time.time()

        # Train
        train_loss = train_epoch(model, train_loader, optimizer, criterion, device, epoch)

        # Validate
        val_loss, val_metrics = validate(model, val_loader, criterion, device, epoch, num_classes)

        epoch_time = time.time() - start_time
        print(f"Epoch {epoch+1}/{epochs} - {epoch_time:.1f}s - "
              f"Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, "
              f"Val Dice: {val_metrics['dice']:.4f}, "
              f"Val VS: {val_metrics['volume_similarity']:.4f}")

        # Save best model
        if val_metrics['dice'] > best_val_dice:
            best_val_dice = val_metrics['dice']
            torch.save(model.state_dict(), save_path)
            print(f"Model saved with Dice score: {best_val_dice:.4f}")

        if (epoch + 1) % 5 == 0:
            checkpoint_path = f"/content/drive/MyDrive/4460_Project/checkpoint_epoch_{epoch+1}.pth"
            torch.save(model.state_dict(), checkpoint_path)
            print(f"Checkpoint saved at epoch {epoch+1}")

        metrics_log["epoch"].append(epoch + 1)
        metrics_log["train_loss"].append(train_loss)
        metrics_log["val_loss"].append(val_loss)
        metrics_log["val_dice"].append(val_metrics["dice"])
        metrics_log["val_vs"].append(val_metrics["volume_similarity"])

    df_log = pd.DataFrame(metrics_log)
    df_log.to_csv("/content/drive/MyDrive/4460_Project/training_log.csv", index=False)
    print("Training log saved to training_log.csv")

    # 🧪 Final Test
    model.load_state_dict(torch.load(save_path))
    test_metrics = test(model, test_loader, device, num_classes)
    print(f"Test Results - Dice: {test_metrics['dice']:.4f}, VS: {test_metrics['volume_similarity']:.4f}")

    return test_metrics
```

```python
import os
import torch
import numpy as np
from medpy.metric.binary import dc, assd
from sklearn.metrics import confusion_matrix
from models.hilbert import decode, encode
from models.MedSegMamba import MedSegMamba

# ============== CONFIG ==============
patch_folder = "/content/drive/MyDrive/4460_Project/hsf_IXI_patch_pt"
label_folder = "/content/drive/MyDrive/4460_Project/hsf_IXI_patch_label"
model_path = "/content/drive/MyDrive/4460_Project/best_model_1.pth"
save_dir = "/content/drive/MyDrive/4460_Project/model_evaluation_IXI"

os.makedirs(save_dir, exist_ok=True)
os.makedirs(os.path.join(save_dir, "metrics"), exist_ok=True)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def load_model(output_ch, model_path):
    model = MedSegMamba(
        img_dim=96,
        patch_dim=8,
        img_ch=1,
        output_ch=output_ch,
        channel_sizes=[1, 32, 64, 128, 256, 1024],
        mamba_d_state=64,
        num_layers=9,
        vss_version='v5',
        mlp_dropout_rate=0,
        attn_dropout_rate=0,
        drop_path_rate=0,
        ssm_expansion_factor=1,
        id=False,
        preact=False,
        maxpool=True,
        upsample=True,
        full_final_block=True,
        scan_type='scan'
    )
    model.load_state_dict(torch.load(model_path, map_location=device))
    return model.to(device).eval()

# ============== LABEL MAPPING ==============

# Unified 13-class label map (based on observed remapped values)
HSF_LABELS = {
    0: "background",
    1: "parasubiculum",
    2: "presubiculum",
    3: "subiculum",
    4: "CA1",
    5: "CA3",
    6: "CA4",
    7: "GC-DG",
    8: "HATA",
    9: "fimbria",
    10: "ML-DG",
    11: "hippocampal_fissure",
    12: "HP_tail"
}

# Convert FreeSurfer label to class index 0–12
def map_fs_hsf_labels(volume):
    label_map = {v: i for i, v in enumerate(HSF_LABELS.keys())}
    return np.vectorize(lambda x: label_map.get(x, 0))(volume)

# ============== EVALUATION METRICS ==============
def compute_metrics(pred, target, num_classes):
    metrics = {
        'dice': np.zeros(num_classes),
        'vs': np.zeros(num_classes),
        'assd': np.zeros(num_classes)
    }

    for c in range(1, num_classes):  # Skip background class 0
        gt_mask = (target == c)
        pred_mask = (pred == c)

        if np.sum(gt_mask) == 0 or np.sum(pred_mask) == 0:
            metrics['dice'][c] = np.nan
            metrics['vs'][c] = np.nan
            metrics['assd'][c] = np.nan
            continue

        # Dice
        metrics['dice'][c] = dc(pred_mask, gt_mask)

        # Volume Similarity
        gt_volume = np.sum(gt_mask)
        pred_volume = np.sum(pred_mask)
        metrics['vs'][c] = 1.0 - abs(gt_volume - pred_volume) / (gt_volume + pred_volume)

        # ASSD
        try:
            metrics['assd'][c] = assd(pred_mask, gt_mask)
        except:
            metrics['assd'][c] = np.nan

    return metrics

# ============== CONFUSION MATRIX ==============
def compute_confusion_matrix(pred, target, num_classes):
    """
    Compute voxel-level confusion matrix between predicted and target label maps.
    """
    pred_flat = pred.flatten()
    target_flat = target.flatten()
    return confusion_matrix(target_flat, pred_flat, labels=list(range(num_classes)))


def plot_confusion_matrix(cm, class_names=None, normalize=False, title="Confusion Matrix", save_path=None):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1, keepdims=True)
        cm = np.nan_to_num(cm)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt=".2f" if normalize else "d",
                cmap="Blues", xticklabels=class_names, yticklabels=class_names)
    plt.title(title)
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.tight_layout()

    if save_path:
        plt.savefig(save_path)
        print(f"Confusion matrix saved to {save_path}")
    else:
        plt.show()
```

```python

model = load_model(output_ch=13, model_path=model_path)
all_metrics = []

file_list = sorted([f for f in os.listdir(patch_folder) if f.endswith(".pt")])
total = len(file_list)

overall_conf = np.zeros((13, 13), dtype=np.int32)

for i, fname in enumerate(file_list):
    print(f"[{i+1}/{total}] Processing {fname}")

    x = torch.load(os.path.join(patch_folder, fname)).unsqueeze(0).unsqueeze(0).to(device)
    y_pred = model(x)
    pred_label = torch.argmax(y_pred, dim=1).cpu().numpy().squeeze()

    label_name = fname.replace(".pt", "_label.pt")
    y_true = torch.load(os.path.join(label_folder, label_name)).numpy().astype(int)
    y_true = map_fs_hsf_labels(y_true)

    metrics = compute_metrics(pred_label, y_true, num_classes=13)
    all_metrics.append(metrics)

    conf = compute_confusion_matrix(pred_label, y_true, num_classes=13)
    overall_conf += conf
```

```python
num_classes = 13
total_dice = np.zeros(num_classes)
total_vs = np.zeros(num_classes)
total_assd = np.zeros(num_classes)

for m in all_metrics:
    total_dice += np.nan_to_num(m["dice"])
    total_vs += np.nan_to_num(m["vs"])
    total_assd += np.nan_to_num(m["assd"])

avg_dice = total_dice / len(all_metrics)
avg_vs = total_vs / len(all_metrics)
avg_assd = total_assd / len(all_metrics)

for i in range(num_classes):
    print(f"Class {i:2d} ({HSF_LABELS[i]:>20}): "
          f"Dice = {avg_dice[i]:.4f} | "
          f"VS = {avg_vs[i]:.4f} | "
          f"ASSD = {avg_assd[i]:.4f}")

avg_dice_fg = avg_dice[1:]
avg_vs_fg = avg_vs[1:]
avg_assd_fg = avg_assd[1:]


print(f"Foreground Mean Dice: {np.mean(avg_dice_fg):.4f}")
print(f"Foreground Mean VS:   {np.mean(avg_vs_fg):.4f}")
print(f"Foreground Mean ASSD: {np.mean(avg_assd_fg):.4f}")
```
